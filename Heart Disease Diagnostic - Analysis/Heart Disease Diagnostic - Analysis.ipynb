## There are fourteen features in Dataset
 
 - age :  The person's age in years
 
 
 - sex: The person's sex (1 = male, 0 = female)
 
 
 - cp: chest pain type
        -- Value 1: typical angina
        -- Value 2: atypical angina
        -- Value 3: non-anginal pain
        -- Value 4: asymptomatic
        
        
 - trestbps: resting blood pressure (in mm Hg on admission to the hospital)
 
 
 - chol: serum cholestoral in mg/dl
 
 
 - fbs: (fasting blood sugar > 120 mg/dl) (1 = true; 0 = false)
 
 
 - restecg: resting electrocardiographic results
        -- Value 0: normal
        -- Value 1: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
        -- Value 2: showing probable or definite left ventricular hypertrophy by Estes' criteria
        
        
 - thalach: maximum heart rate achieved
 
 
 - exang: exercise induced angina (1 = yes; 0 = no)
 
 
 - oldpeak = ST depression induced by exercise relative to rest
 
 
 - slope: the slope of the peak exercise ST segment
        -- Value 1: upsloping
        -- Value 2: flat
        -- Value 3: downsloping
        
        
 -  ca: number of major vessels (0-3) colored by flourosopy
 
 
 -  thal: 3 = normal; 6 = fixed defect; 7 = reversable defect
 
 
 - num: diagnosis of heart disease (angiographic disease status)
        -- Value 0: < 50% diameter narrowing (0=no)
        -- Value 1: > 50% diameter narrowing (1=yes)

## Percentage of people having Heart Disease

```python
num = df_copy.groupby('num').size()
num
```

```python
# Converting Numerical Data into Categorical data

def heart_disease(row):
    if row == 0:
        return 'Negative'
    elif row == 1:
        return 'Positive'
```

```python
# Applying Converted Data into our Dataset with New Column -Heart_Disease

df_copy['Heart_disease'] = df_copy['num'].apply(heart_disease)
```

```python
#Converting Numerical Data into Categorical Data

def gender(row):
    if row==1:
        return 'Male'
    elif row==0:
        return 'Female'
```

```python
#Applying converted data into our dataset with new column - Gender

df_copy['Gender'] = df_copy['sex'].apply(gender)
df_copy.head()
```

```python
#Converting Numerical Data into Categorical Data

def age_range(row):
    if row>=28 and row<37:
        return 'Young Age'
    elif row>=37 and row<47:
        return 'Middle Age'
    elif row>=47:
        return 'Elder Age'
```

```python
df_copy['Age_Range']=df_copy['age'].apply(age_range)
df_copy.head()
```
